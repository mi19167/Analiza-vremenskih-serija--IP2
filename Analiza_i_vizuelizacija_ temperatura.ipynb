```python
godine = df['godina'].unique()
print(godine)
```

```python
df['godina'].fillna(2017, inplace=True)
```

```python
df['godina'].isna().any()
```

```python
nepoznate_vrednosti = df[df.isna().any(axis=1)]
print(nepoznate_vrednosti)
```

```python
df_filtered = df[df['godina'] != 1887]
```

```python
prosecne_vrednosti_po_mesecima = df_filtered.mean()
```

```python
print(prosecne_vrednosti_po_mesecima)
```

```python
df.loc[df['godina'] == 1887] = df.loc[df['godina'] == 1887].fillna(prosecne_vrednosti_po_mesecima)
```

```python
podaci_1887 = df[df['godina'] == 1887]
print(podaci_1887)
```

```python
df_filtered = df[df['godina'] == 1887]

# Izračunavanje prosečne temperature za 1887. godinu
prosecna_temp_1887 = df_filtered[['jan', 'feb', 'mar', 'apr', 'maj', 'jun', 'jul', 'avg', 'sep', 'okt', 'nov', 'dec']].mean().mean()

# Ispisivanje prosečne temperature za 1887. godinu
print(prosecna_temp_1887)
```

```python
import matplotlib.pyplot as plt

# Podaci za graf
godine = df['godina']
prosecne_temperature = df['prosek']

# Kreiranje grafa
plt.figure(figsize=(10, 6))
plt.plot(godine, prosecne_temperature, marker='o', linestyle='-')
plt.title('Prosečna temperatura po godinama')
plt.xlabel('Godina')
plt.ylabel('Prosečna temperatura')
plt.grid(True)
plt.xticks(godine[::5])  # Postavljanje oznaka na svakih pet godina
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('prosecne_temperature_po_godinama.png')

# Prikazivanje grafa
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Podaci za grafik
meseci = ['jan', 'feb', 'mar', 'apr', 'maj', 'jun', 'jul', 'avg', 'sep', 'okt', 'nov', 'dec']
prosecne_temperature_po_mesecima = df[meseci].mean()

# Kreiranje grafika
plt.figure(figsize=(12, 6))
plt.plot(meseci, prosecne_temperature_po_mesecima, marker='o')
plt.title('Prosečna temperatura po mesecima za sve godine')
plt.xlabel('Mesec')
plt.ylabel('Prosečna temperatura')
plt.grid(True)
plt.xticks(rotation=45)

# Postavljanje intervala na y-osi
plt.gca().yaxis.set_major_locator(MultipleLocator(2))
plt.tight_layout()

plt.savefig('prosecne_temperature_po_mesecima.png')
# Prikazivanje grafika
plt.show()
```

```python
import matplotlib.pyplot as plt

# Kolone koje predstavljaju sezone
sezone = ['zima', 'proleće', 'leto', 'jesen']

# Izračunavanje prosečnih temperatura po sezonama
prosecne_temperature_po_sezonama = df[sezone].mean()

# Kreiranje grafikona
plt.figure(figsize=(10, 6))
plt.bar(sezone, prosecne_temperature_po_sezonama)
plt.title('Prosečna temperatura po sezonama za sve godine')
plt.xlabel('Sezona')
plt.ylabel('Prosečna temperatura')
plt.grid(True)
plt.gca().yaxis.set_major_locator(MultipleLocator(2))
plt.tight_layout()
plt.savefig('prosecne_temperature_po_sezonama.png')
# Prikazivanje grafikona
plt.show()
```

```python
import matplotlib.pyplot as plt

# Podaci za grafikon
godine = df['godina']
tsr_amp_mes = df['tsr-amp-mes']
tsr_leto_zima = df['tsr leto-zima']

# Kreiranje grafikona
plt.figure(figsize=(12, 6))
plt.plot(godine, tsr_amp_mes, marker='o', label='TSR Amp Mes')
plt.plot(godine, tsr_leto_zima, marker='o', label='TSR Leto-Zima')
plt.title('TSR Amp Mes i TSR Leto-Zima po godinama')
plt.xlabel('Godina')
plt.ylabel('Vrednost')
plt.legend()
plt.grid(True)
plt.xticks(godine[::5])  # Postavljanje oznaka na svakih pet godina
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('TSR_po_godinama.png')
# Prikazivanje grafikona
plt.show()
```

```python
import matplotlib.pyplot as plt

# Izdvajanje minimalnih i maksimalnih temperatura po godinama
minimalne_temperature_po_godinama = df.groupby('godina').min().min(axis=1)
maksimalne_temperature_po_godinama = df.groupby('godina').max().max(axis=1)


plt.figure(figsize=(12, 6))
plt.plot(minimalne_temperature_po_godinama.index, minimalne_temperature_po_godinama, marker='o', label='Minimalna temperatura')
plt.plot(maksimalne_temperature_po_godinama.index, maksimalne_temperature_po_godinama, marker='o', label='Maksimalna temperatura')
plt.title('Minimalna i maksimalna temperatura po godinama')
plt.xlabel('Godina')
plt.ylabel('Temperatura')
plt.legend()
plt.grid(True)
plt.xticks(godine[::5])  
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('min_i_max_temperature.png')
# Prikazivanje grafikona
plt.show()
```

```python
import matplotlib.pyplot as plt

# Izdvajanje minimalnih i maksimalnih temperatura po mesecima
minimalne_temperature_po_mesecima = df[['jan', 'feb', 'mar', 'apr', 'maj', 'jun', 'jul', 'avg', 'sep', 'okt', 'nov', 'dec']].min()
maksimalne_temperature_po_mesecima = df[['jan', 'feb', 'mar', 'apr', 'maj', 'jun', 'jul', 'avg', 'sep', 'okt', 'nov', 'dec']].max()


meseci = ['jan', 'feb', 'mar', 'apr', 'maj', 'jun', 'jul', 'avg', 'sep', 'okt', 'nov', 'dec']
plt.figure(figsize=(12, 6))
plt.plot(meseci, minimalne_temperature_po_mesecima, marker='o', label='Minimalna temperatura')
plt.plot(meseci, maksimalne_temperature_po_mesecima, marker='o', label='Maksimalna temperatura')
plt.title('Minimalna i maksimalna temperatura po mesecima')
plt.xlabel('Mesec')
plt.ylabel('Temperatura')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grupisanje po godinama i izračunavanje prosečnih temperatura za svako godišnje doba
yearly_seasonal_avg_temperature = df.groupby('godina')[['proleće', 'leto', 'jesen', 'zima']].mean().reset_index()

# Preoblikovanje DataFrame-a radi lakšeg rukovanja
yearly_seasonal_avg_temperature_melted = yearly_seasonal_avg_temperature.melt(id_vars='godina', var_name='Doba', value_name='Prosečna temperatura')

# Vizualizacija
plt.figure(figsize=(14, 8))
palette = sns.color_palette("husl", 4)  # Palette boja za 4 doba
sns.lineplot(data=yearly_seasonal_avg_temperature_melted, x='godina', y='Prosečna temperatura', hue='Doba', palette=palette, marker='o')
plt.title('Prosečne temperature po godinama i doba')
plt.xlabel('Godina')
plt.ylabel('Prosečna temperatura')
plt.legend(title='Doba')
plt.grid(True)
plt.xticks(godine[::5])  # Postavljanje oznaka na svakih pet godina
plt.savefig('prosecne_temperature_godine_sezone.png')
# Prikazivanje grafikona
plt.show()
```
